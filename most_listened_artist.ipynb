Necessary Imports:

1.   pandas: For data analysis
2.   matplotlib: For visualization



```python
import pandas as pd
import matplotlib.pyplot as plt
```

Data Loading Function:

1.   Reads JSON file
2.   Converts time in milliseconds to minutes



```python
def load_data(file_path):
    df = pd.read_json(file_path)
    df['minutes_played'] = df['msPlayed'] / (1000 * 60)
    return df
```

Graphing Function:


1.   Uses Seaborn style
2.   Creates 10x6 graph
3.   Calculate total listening times by artist
4.   Selects top 10 artists
5.   Draws horizontal bar graph










```python
def plot_top_artists(df):
    plt.style.use('seaborn-v0_8')
    plt.figure(figsize=(10, 6))

    top_artists = df.groupby('artistName')['minutes_played'].sum().sort_values(ascending=False).head(10)
    top_artists.plot(kind='barh')
    plt.title('Top 10 Artists by Listening Time')

    plt.tight_layout()
    plt.show()
```

Main Function:



1.   Loads data with error handling
2.   Creates the graph
3.   Calls main() function when script is run directly




```python
def main():
    try:
        df = load_data("StreamingHistory_music_0.json")
        plot_top_artists(df)
    except Exception as e:
        print(f"An error occurred: {str(e)}")

if __name__ == "__main__":
    main()
```
